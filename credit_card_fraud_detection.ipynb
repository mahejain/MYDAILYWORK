```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline as SkPipeline
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_auc_score,
    average_precision_score,
    precision_recall_curve,
    f1_score
)

train_df = pd.read_csv("/kaggle/input/fraud-detection/fraudTrain.csv")
test_df = pd.read_csv("/kaggle/input/fraud-detection/fraudTest.csv")

# DROP LESS USEFUL / ID-LIKE COLUMNS
drop_cols = ["Unnamed: 0", "cc_num", "first", "last", "street", "trans_num"]
train_df.drop(columns=drop_cols, inplace=True, errors="ignore")
test_df.drop(columns=drop_cols, inplace=True, errors="ignore")

# FEATURE ENGINEERING
def engineer_features(df):
    df = df.copy()

    df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])
    df["hour"] = df["trans_date_trans_time"].dt.hour
    df["is_night"] = df["hour"].isin([0,1,2,3,4,5]).astype(int)
    df["day"] = df["trans_date_trans_time"].dt.day
    df["month"] = df["trans_date_trans_time"].dt.month
    df["weekday"] = df["trans_date_trans_time"].dt.weekday
    df["is_weekend"] = df["weekday"].isin([5, 6]).astype(int)

    df["dob"] = pd.to_datetime(df["dob"])
    df["age"] = ((df["trans_date_trans_time"] - df["dob"]).dt.days // 365).clip(lower=0)

    df["amt_log"] = np.log1p(df["amt"])

    if all(col in df.columns for col in ["lat", "long", "merch_lat", "merch_long"]):
        df["merchant_distance"] = np.sqrt(
            (df["lat"] - df["merch_lat"]) ** 2 +
            (df["long"] - df["merch_long"]) ** 2
        )

    df.drop(columns=["trans_date_trans_time", "dob"], inplace=True, errors="ignore")

    return df

train_df = engineer_features(train_df)
test_df = engineer_features(test_df)

# SPLIT FEATURES / TARGET
X_train = train_df.drop("is_fraud", axis=1)
y_train = train_df["is_fraud"]

X_test = test_df.drop("is_fraud", axis=1)
y_test = test_df["is_fraud"]

# IDENTIFY COLUMN TYPES
categorical_cols = X_train.select_dtypes(include=["object"]).columns.tolist()
numerical_cols = X_train.select_dtypes(exclude=["object"]).columns.tolist()

print("Categorical columns:", categorical_cols)
print("Numerical columns:", numerical_cols)

# PREPROCESSORS
numeric_transformer_lr = SkPipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = SkPipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor_lr = ColumnTransformer([
    ("num", numeric_transformer_lr, numerical_cols),
    ("cat", categorical_transformer, categorical_cols)
])

numeric_transformer_rf = SkPipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

preprocessor_rf = ColumnTransformer([
    ("cat", categorical_transformer, categorical_cols),
    ("num", numeric_transformer_rf, numerical_cols)
])

# LOGISTIC REGRESSION MODEL
pipeline_lr = Pipeline([
    ("preprocess", preprocessor_lr),
    ("smote", SMOTE(random_state=42)),
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        random_state=42
    ))
])

pipeline_lr.fit(X_train, y_train)
y_prob_lr = pipeline_lr.predict_proba(X_test)[:, 1]

print("\n=========================")
print("LOGISTIC REGRESSION")
print("=========================")

# Threshold tuning
thresholds = [0.3, 0.5, 0.7, 0.9]

for t in thresholds:
    y_pred_lr_t = (y_prob_lr > t).astype(int)
    print(f"\n--- Logistic Regression Threshold: {t} ---")
    print("Predicted frauds:", y_pred_lr_t.sum())
    print(classification_report(y_test, y_pred_lr_t, zero_division=0))

# Best threshold
best_thresh_lr = 0
best_f1_lr = 0

for t in np.linspace(0.1, 0.9, 17):
    y_pred_lr_t = (y_prob_lr > t).astype(int)
    f1 = f1_score(y_test, y_pred_lr_t, zero_division=0)
    if f1 > best_f1_lr:
        best_f1_lr = f1
        best_thresh_lr = t

print("\nBest Logistic Regression Threshold:", best_thresh_lr)
print("Best Logistic Regression F1:", best_f1_lr)

# PR curve
precision_lr, recall_lr, thresh_lr = precision_recall_curve(y_test, y_prob_lr)

plt.figure(figsize=(8, 5))
plt.plot(thresh_lr, precision_lr[:-1], label="Precision")
plt.plot(thresh_lr, recall_lr[:-1], label="Recall")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Logistic Regression: Precision-Recall vs Threshold")
plt.legend()
plt.show()

# Confusion matrix
y_pred_lr_best = (y_prob_lr > best_thresh_lr).astype(int)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_lr_best)
plt.title("Logistic Regression Confusion Matrix")
plt.show()

print("Logistic Regression ROC-AUC:", roc_auc_score(y_test, y_prob_lr))
print("Logistic Regression PR-AUC:", average_precision_score(y_test, y_prob_lr))

# RANDOM FOREST MODEL
pipeline_rf = Pipeline([
    ("preprocess", preprocessor_rf),
    ("model", RandomForestClassifier(
        n_estimators=100,
        max_depth=15,
        min_samples_split=10,
        min_samples_leaf=2,
        class_weight="balanced_subsample",
        random_state=42,
        n_jobs=-1
    ))
])

pipeline_rf.fit(X_train, y_train)
y_prob_rf = pipeline_rf.predict_proba(X_test)[:, 1]
y_pred_rf_default = (y_prob_rf > 0.5).astype(int)

print("\n=========================")
print("RANDOM FOREST")
print("=========================")

print("\n--- Random Forest Threshold: 0.5 ---")
print("Predicted frauds:", y_pred_rf_default.sum())
print(classification_report(y_test, y_pred_rf_default, zero_division=0))

for t in thresholds:
    y_pred_rf_t = (y_prob_rf > t).astype(int)
    print(f"\n--- Random Forest Threshold: {t} ---")
    print("Predicted frauds:", y_pred_rf_t.sum())
    print(classification_report(y_test, y_pred_rf_t, zero_division=0))

# Best threshold
best_thresh_rf = 0
best_f1_rf = 0

for t in np.linspace(0.1, 0.9, 17):
    y_pred_rf_t = (y_prob_rf > t).astype(int)
    f1 = f1_score(y_test, y_pred_rf_t, zero_division=0)
    if f1 > best_f1_rf:
        best_f1_rf = f1
        best_thresh_rf = t

print("\nBest Random Forest Threshold:", best_thresh_rf)
print("Best Random Forest F1:", best_f1_rf)

# PR curve
precision_rf, recall_rf, thresh_rf = precision_recall_curve(y_test, y_prob_rf)

plt.figure(figsize=(8, 5))
plt.plot(thresh_rf, precision_rf[:-1], label="Precision")
plt.plot(thresh_rf, recall_rf[:-1], label="Recall")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Random Forest: Precision-Recall vs Threshold")
plt.legend()
plt.show()

# Confusion matrix
y_pred_rf_best = (y_prob_rf > best_thresh_rf).astype(int)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf_best)
plt.title("Random Forest Confusion Matrix")
plt.show()

print("Random Forest ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
print("Random Forest PR-AUC:", average_precision_score(y_test, y_prob_rf))

rf_model = pipeline_rf.named_steps["model"]

ohe = pipeline_rf.named_steps["preprocess"].named_transformers_["cat"].named_steps["onehot"]
ohe_features = ohe.get_feature_names_out(categorical_cols)

all_features = list(ohe_features) + list(numerical_cols)

importance = pd.Series(rf_model.feature_importances_, index=all_features).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
importance.head(10).sort_values().plot(kind="barh")
plt.title("Top 10 Important Features (Random Forest)")
plt.xlabel("Importance")
plt.show()

print("\nTop 10 Important Features:")
print(importance.head(10))

print("\n=========================")
print("MODEL COMPARISON")
print("=========================")
print(f"Logistic Regression ROC-AUC: {roc_auc_score(y_test, y_prob_lr):.4f}")
print(f"Logistic Regression PR-AUC:  {average_precision_score(y_test, y_prob_lr):.4f}")
print(f"Random Forest ROC-AUC:       {roc_auc_score(y_test, y_prob_rf):.4f}")
print(f"Random Forest PR-AUC:        {average_precision_score(y_test, y_prob_rf):.4f}")
```
